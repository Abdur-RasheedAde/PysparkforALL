```python
!pip install pyspark py4j
import pyspark
path ="/iris_dataset.csv"
from pyspark.sql import  SparkSession
spark = SparkSession.builder.appName('LetsGO2').getOrCreate()
```

```python
data = '/content/testdata2.csv'
pydf = spark.read.csv(data, header=True, inferSchema=True)
pydf.show()
```

```python
#Dropping any rows with null values
pydf.na.drop().show()
```

```python
#Drop any row with null just like the code above
pydf.na.drop(how='any').show()
```

```python
#Don't delete records with at least 2 non-null values
pydf.na.drop(how='any', thresh=2).show()
```

```python
#Subset: delete records from a specific column
pydf.na.drop(how='any', subset=['Experience']).show()
```

```python
#Fill missing values with the word 'Missing values'
pydf.na.fill('Missing Values').show()
```

```python
#Fill missing values with the word 'Missing values' in 2 columns (NW)
pydf.na.fill('Missing Values',['Experience','age']).show()
```

```python
pydf.printSchema()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['Age', 'Experience', 'Salary'],
    outputCols=["{}_imputed".format(c) for c in ['Age', 'Experience', 'Salary']]
    ).setStrategy("median")
```

```python
# Add imputation cols to df to fill null with means
imputer.fit(pydf).transform(pydf).show()
```

```python
data3 = '/content/testdata3.csv'
pydf2 = spark.read.csv(data3, header=True, inferSchema=True)
pydf2.show()
```

```python
#Salary <=30000
pydf2.filter('Salary<=35000').show()
```

```python
#Salary <=30000 same as above with different method
pydf2.filter(pydf2['Salary']<=35000).show()
```

```python
#Salary <=30000 showing only Name and Age
pydf2.filter('Salary<=35000').select(['Name', 'Age']).show()
```

```python
#Salary <=40000 ABD >30000
pydf2.filter((pydf2['Salary']>30000) & (pydf2['Salary']<=40000)).show()
```

```python
#Salary <=40000 ABD >30000 same as above with SELECT specific column
pydf2.filter((pydf2['Salary']>30000) & (pydf2['Salary']<=40000)).select(['Name', 'Age']).show()
```

```python
#NOT <=30000 using (Inverse operation) '~'
pydf2.filter(~(pydf2['Salary']<=35000)).select(['Name', 'Age', "Salary"]).show()
```

```python
data4 = '/content/testdata4.csv'
pydf4 = spark.read.csv(data4, header=True, inferSchema=True)
pydf4.show()
```

```python
# Group by sum of Salary of each name
pydf4.groupBy('Name').sum().show()
```

```python
# Group by mean of Salary of Department
pydf4.groupBy('Department').mean().show()
```

```python
# Count number of workers in Department
pydf4.groupBy('Department').count().show()
```

```python
# Total sum of all Salary
pydf4.agg({'Salary': 'sum'}).show()
```

```python
# Maximum Salary from each Department
pydf4.groupBy('Department').max().show()
```

```python
# Minimum Salary from each Name
pydf4.groupBy('Name').min().show()
```

```python
# Average Salary from each Name
pydf4.groupBy('Name').avg().show()
```
